```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def update_stock(self, quantity):
        self.stock += quantity

    def __str__(self):
        return f"Product: {self.name}, Price: ${self.price:.2f}, Stock: {self.stock}"

# Store inventory dictionary
products = {}

def add_product():
    name = input("Enter product name: ")
    price = float(input("Enter product price: "))
    stock = int(input("Enter initial stock quantity: "))
    products[name] = Product(name, price, stock)
    print("Product added successfully!")

def update_product_stock():
    name = input("Enter product name to update stock: ")
    if name in products:
        quantity = int(input("Enter quantity to add/remove (negative to remove): "))
        products[name].update_stock(quantity)
        print("Stock updated successfully!")
    else:
        print("Product not found!")

def view_product():
    name = input("Enter product name to view details: ")
    if name in products:
        print(products[name])
    else:
        print("Product not found!")

def main():
    while True:
        print("\nStore Product Manager")
        print("1. Add Product")
        print("2. Update Product Stock")
        print("3. View Product Details")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_product()
        elif choice == '2':
            update_product_stock()
        elif choice == '3':
            view_product()
        elif choice == '4':
            print("Exiting program. Goodbye!")
            break


if __name__ == "__main__":
    main()
```
